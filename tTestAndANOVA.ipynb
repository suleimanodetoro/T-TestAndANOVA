# Part 1: Parametric Tests (e.g T-Test) – Assumption: Dataset has a normal distribution (Sometimes, you might assume things like Homogeneity of Variance (for Independent Two-Sample t-test): This assumption states that the variances of the two groups being compared should be equal. The Levene's test can be used to check this assumption. If this assumption is violated, a variation of the t-test called Welch's t-test might be used.

Independence: Observations should be independent of each other. This means that the occurrence of one observation should not affect the occurrence of another observation.

Scale of Measurement: The dependent variable should be measured on a continuous or ordinal scale.

Random Sampling: The data should be derived from a random sample from the population.)

The comment here indicates that the tests you're about to perform assume that the dataset follows a normal distribution. Parametric tests often make certain assumptions about the properties of the data, and in this case, the assumption is normality.

T-Test
This is a type of hypothesis test used to determine if there's a significant difference between the means of two groups.

Type 1: One-Sample T-test
The one-sample t-test is used to determine whether a sample comes from a population with a specific mean. Essentially, you're comparing the mean of a single sample of scores to a known or hypothesized value to see if they are statistically significantly different.

```python
# Importing necessary libraries
import pandas as pd
from scipy import stats

# Loading the dataset
dataset = pd.read_csv('brain_size.csv', sep=';', na_values=".")

# Performing the one-sample t-test for VIQ against a population mean of 0
t_statistic, p_value = stats.ttest_1samp(dataset['VIQ'], 0)

# Displaying the results
print(f"Ttest_1sampResult(statistic={t_statistic:.3f}, pvalue={p_value:.5e})\n")

# Interpretation
print("Interpretation of the Output:")
print(f"t-value = {t_statistic:.3f}")
print(f"p-value = {p_value:.5e} (this means {p_value} x 10 to the power of -28)\n")

# Setting the confidence levels (could this be known as significance level too?) (alpha values) for hypothesis testing
alpha_values = [0.05, 0.02, 0.01]

# Defining the hypotheses
print("Null Hypothesis (H0): mean of VIQ = 0;")
print("Alternative Hypothesis (Ha): mean of VIQ ≠ 0\n")

# Interpretation based on different alpha values
for alpha in alpha_values:
    print(f"Choosing Confidence Level of {(1-alpha)*100}% or alpha value, α = {alpha}:")
    if p_value < alpha / 2:  # Two-tailed test
        print(f"Since p-value = {p_value:.5e} < α/2 = {alpha/2}, we reject the null hypothesis.")
        print(f"Therefore, the conclusion is: mean of VIQ is not equal to 0.\n")
    else:
        print(f"Since p-value = {p_value:.5e} >= α/2 = {alpha/2}, we fail to reject the null hypothesis.")
        print(f"Therefore, the conclusion is: mean of VIQ is equal to 0.\n")
```

# The next cell explore an example for sameple t-test
The Wechsler Adult Intelligence Scale and the Stanford-Binet test, scores that fall between 90 and 109 are considered average IQ scores.
Choose a 2-tail t-test the following hypotheses:
Null Hypothesis (H0): mean of VIQ = 100;
Alternative Hypothesis (Ha): mean of VIQ  ≠ 100

Run the above hypothesis testing and provide an interpretation of the results for  = 0.05; 0.02; 0.01.


```python
# Importing the required module from scipy for statistical tests
from scipy import stats

# Perform a one-sample t-test on the 'VIQ' column of the dataset.
# The goal is to compare the mean VIQ score from our dataset to a known value (in this case, 100)
# to see if there's a statistically significant difference.
t_statistic, p_value = stats.ttest_1samp(dataset['VIQ'], 100)

# List of significance levels (alpha values) for hypothesis testing.
# These values represent the probability of rejecting the null hypothesis when it's actually true.
alpha_values = [0.05, 0.02, 0.01]

# An empty list to store the interpretations of the t-test results for each alpha value.
interpretations = []

# Loop through each alpha value to interpret the t-test results.
for alpha in alpha_values:
    
    # Check if the p-value is less than the significance level (alpha).
    # If it is, it suggests that the observed data is inconsistent with the null hypothesis.
    if p_value < alpha:
        result = f"For α = {alpha}: Reject H0. Evidence suggests the mean of VIQ is different from 100."
        
    # If the p-value is greater than or equal to alpha, we don't have enough evidence to reject the null hypothesis.
    else:
        result = f"For α = {alpha}: Fail to reject H0. Insufficient evidence to say the mean of VIQ is different from 100."
    
    # Append the interpretation to our list.
    interpretations.append(result)

# Output the t-statistic, p-value, and interpretations.
t_statistic, p_value, interpretations
```

```python
# Importing necessary libraries
import pandas as pd
from scipy import stats

# Loading the dataset
dataset = pd.read_csv('brain_size.csv', sep=';', na_values=".")

# Extracting FSIQ values for females and males
female_fsiq = dataset[dataset['Gender'] == 'Female']['FSIQ']
male_fsiq = dataset[dataset['Gender'] == 'Male']['FSIQ']

# Performing the independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(female_fsiq, male_fsiq)

# Displaying the results
print(f"Ttest_indResult(statistic={t_statistic:.3f}, pvalue={p_value:.3f})\n")

# Interpretation
print("Interpretation of the Output:")
print(f"t-value = {t_statistic:.3f}")
print(f"p-value = {p_value:.3f}\n")

# Setting the significance levels (alpha values) for hypothesis testing
alpha_values = [0.05, 0.02, 0.01]

# Defining the hypotheses
print("Null Hypothesis (H0): Female FSIQ mean = Male FSIQ mean;")
print("Alternative Hypothesis (Ha): Female FSIQ mean ≠ Male FSIQ mean\n")

# Interpretation based on alpha = 0.01
alpha = 0.01
print(f"Choosing Confidence Level of {(1-alpha)*100}% or alpha value, α = {alpha}:")
if p_value < alpha:
    print(f"Since p-value = {p_value:.3f} < α = {alpha}, we reject the null hypothesis.")
    print(f"Therefore, the conclusion is: there is a significant difference between the Female and Male FSIQ means.")
else:
    print(f"Since p-value = {p_value:.3f} >= α = {alpha}, we fail to reject the null hypothesis.")
    print(f"Therefore, the conclusion is: there is no significant difference between the Female and Male FSIQ means.")
```

# Next Cell: Example 2.2 for 2-sample t-test
Choose a 2-tail t-test the following hypotheses:
Null Hypothesis (H0): Female PIQ mean = Male PIQ mean;
Alternative Hypothesis (Ha): Female PIQ mean  Male ≠ PIQ mean;

Run the above hypothesis testing and provide an interpretation of the results for  = 0.05; 0.02; 0.01.


```python
# Importing necessary libraries
import pandas as pd
from scipy import stats

# Loading the dataset
dataset = pd.read_csv('brain_size.csv', sep=';', na_values=".")

# Extracting PIQ values for females and males
female_piq = dataset[dataset['Gender'] == 'Female']['PIQ']
male_piq = dataset[dataset['Gender'] == 'Male']['PIQ']

# Performing the independent two-sample t-test
t_statistic_piq, p_value_piq = stats.ttest_ind(female_piq, male_piq)

# Displaying the results
print(f"Ttest_indResult(statistic={t_statistic_piq:.3f}, pvalue={p_value_piq:.5f})\n")

# Setting the significance levels (alpha values) for hypothesis testing
alpha_values = [0.05, 0.02, 0.01]

# Defining the hypotheses
print("Null Hypothesis (H0): Female PIQ mean = Male PIQ mean;")
print("Alternative Hypothesis (Ha): Female PIQ mean ≠ Male PIQ mean\n")

# Interpretation based on different alpha values
for alpha in alpha_values:
    if p_value_piq < alpha:
        print(f"For α = {alpha}: Reject H0. Evidence suggests the means of Female and Male PIQ are different.\n")
    else:
        print(f"For α = {alpha}: Fail to reject H0. Insufficient evidence to say the means of Female and Male PIQ are different.\n")
```

# One Way Analysis of Variance (ANOVA)
Assumptions for ANOVA test:
    1. The samples are independent
    2. Each Sample is from a normally distributted population
    3. The population standard deviations of the groups are all equal. The property is known as homoscendasticity
    

```python
# 	Descriptive statistics and boxplot for the dataset
import pandas as pd
from matplotlib import pyplot as plt

dataset = pd.read_excel("score.xlsx")
print(dataset.describe())

dataset.boxplot(column=['A', 'B', 'C', 'D'], grid=False)
plt.title("Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Score")

```

```python
# import pandas as pd
import scipy.stats as stats

# Load the dataset
dataset_score = pd.read_excel("score.xlsx")

# Perform the one-way ANOVA
# We would like to compare the means of Subjects A, B, C, and D. The two hypotheses are as follows:
# Note: for ANOVA, it is only one-tail test
# Null Hypothesis (H0): A mean = B mean = C mean = D mean (all the means of all the subjects are equal to each other);
# Alternative Hypothesis (Ha): At least one of the means for A, B, C, and D is not equal to the other

fvalue, pvalue = stats.f_oneway(dataset_score["A"], dataset_score["B"], dataset_score["C"], dataset_score["D"])

# Display the ANOVA results
print(f"F Value  = {fvalue:.6f}")
print(f"P Value  = {pvalue:.6e}")

# Set the confidence (significance) level. We can also use 
alpha = 0.05

# Automatically draw the conclusion based on the p-value
if pvalue < alpha:
    print("The null hypothesis can be rejected")
else:
    print("The null hypothesis is accepted")
```
